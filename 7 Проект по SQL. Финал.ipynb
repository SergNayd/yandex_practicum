## Для каждой книги считаем кол-во обзоров и среднюю оценку:

```python
query_2 = ''' 

SELECT 
    books.title AS book_name,
    books.book_id AS id,
    SubRev.n_rev AS reviews,
    SubRat.avg_rat AS avg_rating
FROM books LEFT JOIN 
                (SELECT
                    book_id,
                    AVG(rating) AS avg_rat
                FROM
                    ratings
                GROUP BY 
                    book_id
                ) AS SubRat ON SubRat.book_id = books.book_id
        LEFT JOIN
                (SELECT 
                    book_id,
                    COUNT(review_id) AS n_rev
                FROM
                    reviews
                GROUP BY 
                    book_id
                ) AS SubRev ON SubRev.book_id = books.book_id
                
; ''' 

book_review = pd.io.sql.read_sql(query_2, con = engine)
book_review
```

### Предварительный вывод:
Таблица с подсчётом на каждую книгу - кол-во обзоров и средня оценка готова

## Определим издательство, которое выпустило наибольшее число книг толще 50 страниц — так мы исключим из анализа брошюры:

```python
query_3 = ''' 
SELECT
    T.publisher_id,
    T.publisher,
    T.n_books
FROM
    (SELECT 
        books.publisher_id AS publisher_id,
        COUNT(books.book_id) AS n_books,
        publishers.publisher AS publisher 
    FROM 
        books 
        LEFT JOIN publishers ON books.publisher_id=publishers.publisher_id

    WHERE 
        num_pages>50
    GROUP BY
        books.publisher_id,
        publishers.publisher
    ORDER BY 
        n_books DESC) AS T
LIMIT 1
; ''' 

book_volume = pd.io.sql.read_sql(query_3, con = engine)
book_volume
```

### Предварительный вывод:
`Penguin Books` - издательство, которое выпустило наибольшее число книг толще 50 страниц : 42 книги.

## Определим автора с самой высокой средней оценкой книг — учитываем только книги с 50 и более оценками:

```python
query_4 = ''' 

SELECT 
    authors.author AS author
   , AVG(ratings.rating) AS avg_rating
FROM books
    LEFT JOIN authors ON books.author_id = authors.author_id
    LEFT JOIN ratings ON books.book_id = ratings.book_id
GROUP BY 
     books.book_id
     , authors.author_id
HAVING
    COUNT(ratings.rating_id) >=50
ORDER BY 
    AVG(ratings.rating) DESC
LIMIT 5
    
; ''' 


author = pd.io.sql.read_sql(query_4, con = engine)
author
```

### Предварительный вывод:
Автор с самой высокой средней оценкой книг - `J.K. Rowling/Mary GrandPré`. Средний рейтинг 4,41

## Посчитаем среднее кол-во обзоров от пользователей, которые поставили больше 50 оценок:


```python
query_5 = ''' 

SELECT
    AVG(Sub.n_reviews)
FROM
    (SELECT
        username,
        COUNT(review_id) AS n_reviews
    FROM
        reviews
    WHERE 
        username IN (
                SELECT 
                    username
                FROM 
                    ratings
                GROUP BY 
                    username
                HAVING 
                    COUNT(rating_id) > 50
                    )
    GROUP BY 
        username
    ) AS Sub
; ''' 

user_overview = pd.io.sql.read_sql(query_5, con = engine)
user_overview
```

### Предварительный вывод:
Среднее количество обзоров от пользователей, которые поставили больше 50 оценок: 24,33
